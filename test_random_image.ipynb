```python
import cv2
from  keras.models import load_model
from PIL import Image
import numpy as np
import os
```

```python
model = load_model("D:\College\Capstone Project 2\data\Trained_models\Big_Brain_tumor_model_1.h5")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
def plot_image(path):
    img = mpimg.imread(path)
    imgplot = plt.imshow(img)
    plt.show()
```

```python
path = r"D:\College\Capstone Project 2\data\brain_tumor_dataset_testing\mixed_pred\t33.jpg"
```

```python
image = Image.open(path)
image = image.resize((224,224))
image = image.convert('RGB')
image.save(path)
image = cv2.imread(path)
img = np.array(image)
img = np.expand_dims(img, axis = 0)
result = model.predict(img)

if result.squeeze() >= 0.5:
    print(f"Brain Tumor")
else:
    print("No Brain Tumor")
# print(result)
```

```python
plot_image(path)
```
